## 折线图

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,101)
x
```

```python
siny = np.sin(x)
cosy = np.cos(x)
```

```python
%matplotlib inline
plt.title('Sin graph')
plt.xlabel('X')
plt.ylabel('Y')
plt.ylim(-2,2)
plt.plot(x, siny, color = '#66ccff', label='sin(x)')
plt.plot(x, cosy, linestyle = ':', label='cos(x)')
plt.legend()
plt.show()
```

## 散点图

```python
x = np.linspace(0,10,10001)
y = np.random.normal(0,10,10001) # 正太分布
plt.scatter(x,y, alpha=0.6) # 绘制散点图，alpha 为不透明度
plt.show()
```
